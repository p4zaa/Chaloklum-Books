# bookList
Provide books detail corressponding to id

```python
bookList = df[['id', 'title','subtitle', 'authors', 'language', 'genre', 'published_date', 'page_count']]
```

```python
for index in bookList.index:
  authors_str = df.at[index, 'authors']
  genres_str = df.at[index, 'genre']

  ### Authors list split ###
  if authors_str is not np.nan:
    if type(authors_str) is str:
      authors_list = ast.literal_eval(authors_str)
      bookList.at[index, 'authors'] = ' | '.join(str(e) for e in authors_list)
    else:
      bookList.at[index, 'authors'] = ' | '.join(str(e) for e in authors_str)

  ### Genres list split ###
  if genres_str is not np.nan:
    if type(genres_str) is str:
      genres_list = ast.literal_eval(genres_str)
      bookList.at[index, 'genre'] = ' | '.join(str(e) for e in genres_list)
    else:
      bookList.at[index, 'genre'] = ' | '.join(str(e) for e in genres_str)
```

# bookReview

```python
id_list = []
rating_list = []
text_list = []

for index in df.index:
  id = df.at[index, 'id']
  rateNreview = df.at[index, 'text_reviews']
  if rateNreview is not np.nan:
    rateNreview = ast.literal_eval(rateNreview)
    for review in rateNreview:
      rating = review[0]
      text = review[1]
      id_list.append(id)
      rating_list.append(rating)
      text_list.append(text)
```

```python
bookReview = pd.DataFrame()
bookReview = bookReview.assign(id = id_list)
bookReview = bookReview.assign(rating = rating_list)
bookReview = bookReview.assign(text_review = text_list)
```

```python
bookReview
```

# bookAlsoEnjoy

```python
bookAlsoEnjoy = df[['id', 'also_enjoy']]
```

```python
bookAlsoEnjoy = bookAlsoEnjoy[bookAlsoEnjoy['also_enjoy'].notna()]
```

```python
for index in bookAlsoEnjoy.index:
  book_list = []
  joyset = bookAlsoEnjoy.at[index, 'also_enjoy']
  joyset = ast.literal_eval(joyset)
  for each in joyset:
    book = each[0]
    url = each[1]
    book_list.append(book)
  bookAlsoEnjoy.at[index, 'also_enjoy'] = ' | '.join(str(e) for e in book_list)
```
